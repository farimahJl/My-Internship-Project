```python
import pandas as pd
from PIL import ImageColor
import numpy as np
```

# Code to create the lookup csv

```python
# copied and paste table from 
# https://www.homebrewtalk.com/threads/ebc-or-srm-to-color-rgb.78018/ into flat file

with open('srm_lookup') as f:
    brew = {}
    for line in f:
        no, color = line.split('=')
        brew[no] = color.strip()
    
    data = {'srm': brew.keys(),
            'hex': brew.values()} 
    df = pd.DataFrame.from_dict(data)

df.srm = df.srm.astype(int)

#transfer hex to RGB tuple
df['RGB'] = df['hex'].apply(lambda x: ImageColor.getcolor(x, "RGB"))
#expand into different columns
df.loc[:,'red']=df.RGB.map(lambda x:x[0])
df.loc[:,'green']=df.RGB.map(lambda x:x[1])
df.loc[:,'blue']=df.RGB.map(lambda x:x[2])
```

```python
df.to_csv('lookup_srm.csv', index = False)
```

# Code to lookup srm value based on the lookup table

```python
#https://www.codegrepper.com/tpc/how+to+find+the+closest+color+in+python



def closest_color(r, g, b, lookup_table):
    COLORS = lookup_table['RGB']
    color_diffs = []
    for color in COLORS:
        cr, cg, cb = color
        color_diff = np.sqrt(abs(r - cr)**2 + abs(g - cg)**2 + abs(b - cb)**2)
        color_diffs.append((color_diff, color)) # store diff with color from srm table
        closest_color = min(color_diffs)[1] # search minimum diff and extract color part
    return closest_color

def srm_value(r, g, b, lookup_table):
    cc = closest_color(r, g, b, lookup_table)
    print(f'the closest color of ({r,g,b}) is {cc}')
    #get location from table with closest color and extract srm value
    srm = lookup_table.loc[lookup_table['RGB'] == cc, 'srm'].values[0]
    print(f'the srm value that belongs to closest color is {srm}')
    return srm
```
